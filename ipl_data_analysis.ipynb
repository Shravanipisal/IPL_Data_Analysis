```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))
```

```python
df= pd.read_csv("/content/deliveries (1).csv")
df
```

```python
print(f"Shape:{df.shape}")
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
#data cleaning and feature engineering
print(df.isnull().sum()/ len(df)*100)
```

```python
# Fill missing values in 'player_dismissed' and 'dismissal_kind' with 'Not Out'
df['player_dismissed'] = df['player_dismissed'].fillna('Not Out')
df['dismissal_kind'] = df['dismissal_kind'].fillna('Not Out')
df['fielder'] = df['fielder'].fillna('None')
df
```

```python
# Create a new feature: 'total_runs_scored' (same as total_runs)
df['total_runs_scored'] = df['total_runs']
df['total_runs_scored']
```

```python
# Create a feature: 'is_boundary' (4 or 6 runs from batsman)
df['is_boundary'] = (df['batsman_runs'].isin([4, 6])).astype(int)
df['is_boundary']
```

```python
# Create a feature: 'is_wicket' (if player was dismissed)
df['is_wicket'] = (df['player_dismissed'] != 'Not Out').astype(int)

df['is_wicket']
```

```python
# Create a feature: 'is_dot_ball' (ball with 0 runs and no wicket)
df['is_dot_ball'] = ((df['total_runs'] == 0) & (df['is_wicket'] == 0)).astype(int)

df['is_dot_ball']
```

```python
# Create a feature: 'run_rate' (runs per ball - could be aggregated later)
# We'll create a 'runs_per_ball' column
df['runs_per_ball'] = df['total_runs']
df['runs_per_ball']
```

```python
# Create a feature: 'run_rate' (runs per ball - could be aggregated later)
# We'll create a 'runs_per_ball' column
df['runs_per_ball'] = df['total_runs']
df['runs_per_ball']
```

```python
# Add a 'ball_number' feature: over * 6 + ball (but careful, ball can be >6 due to extras)
# Let's create a simple ball number
df['ball_number'] = (df['over'] * 6) + df['ball']
df['ball_number']
```

```python
print(df[['batsman_runs', 'total_runs', 'is_boundary', 'is_wicket', 'is_dot_ball', 'ball_number']].head())
```

```python
#Exploratory Data Analysis (EDA)

#Distribution of total runs per delivery
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
df['total_runs'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Runs per Ball')
plt.xlabel('Runs')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
df['total_runs'].value_counts(normalize=True).sort_index().plot(kind='bar')
plt.title('Percentage Distribution of Runs per Ball')
plt.xlabel('Runs')
plt.ylabel('Percentage')
plt.show()
```

```python
#Dot balls, boundaries, and wickets
dot_balls = df['is_dot_ball'].sum()
boundaries = df['is_boundary'].sum()
wickets = df['is_wicket'].sum()

print(f"Dot Balls: {dot_balls}")
print(f"Boundaries: {boundaries}")
print(f"Wickets: {wickets}")

plt.bar(['Dot Balls', 'Boundaries', 'Wickets'], [dot_balls, boundaries, wickets], color=['blue', 'green', 'red'])
plt.title('Key Events per Ball')
plt.show()
```

```python
#Top batsmen by runs scored
top_batsmen = df.groupby('batsman')['batsman_runs'].sum().nlargest(10)

# Plot
top_batsmen.plot(kind='bar', color='red')
plt.title('Top 10 Batsmen by Total Runs')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.xticks(rotation=90)
plt.show()

# Display top 5
print("\nTop 5 Batsmen:")
print(top_batsmen.head())
```

```python
#Top bowlers by wickets taken
top_bowlers = df[df['is_wicket'] == 1].groupby('bowler')['is_wicket'].sum().nlargest(10)


top_bowlers.plot(kind='bar', color='blue')
plt.title('Top 10 Bowlers by Wickets Taken')
plt.xlabel('Bowler')
plt.ylabel('Wickets')
plt.xticks(rotation=90)
plt.show()


print("\nTop 5 Bowlers:")
print(top_bowlers.head())
```

```python
#Team performance: Total runs per team
team_runs = df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)


team_runs.plot(kind='bar', color='gold', figsize=(12, 6))
plt.title('Total Runs Scored by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

print("\nTeam Runs Totals:")
print(team_runs)
```

```python
#extra run analysis
extra_types= ['wide_runs','bye_runs','legbye_runs','noball_runs','penalty_runs']
# Calculate extra runs
extra_totals = df[extra_types].sum().sort_values(ascending=False)


extra_totals.plot(kind='bar', color='purple', figsize=(10, 6))
plt.title('Distribution of Extra Runs by Type')
plt.xlabel('Extra Type')
plt.ylabel('Total Extra Runs')
plt.xticks(rotation=0)
plt.show()


print("\nExtra Runs Breakdown:")
print(extra_totals)
print(f"\nTotal Extra Runs: {extra_totals.sum()}")
```

```python
#Average runs per over (across all matches)
# Calculate runs per over
runs_per_over = df.groupby('over')['total_runs'].mean() * 6

plt.figure(figsize=(14, 6))
runs_per_over.plot(kind='bar', color='black')
plt.title('Average Runs per Over')
plt.xlabel('Over')
plt.ylabel('Runs per Over')
plt.show()
```

```python
# Group by match, inning, and batting team to see total runs
top_teams = df.groupby(['batting_team', 'match_id'])['total_runs'].sum().nlargest(10)

print("\nTop 10 Team Scores:")
print(top_teams)
```

```python
# Get top 5 batsmen and bowlers
top_batsmen = df.groupby('batsman')['batsman_runs'].sum().nlargest(5).index
top_bowlers = df[df['is_wicket'] == 1]['bowler'].value_counts().head(5).index

# Create and plot heatmap
df_filtered = df[df['batsman'].isin(top_batsmen) & df['bowler'].isin(top_bowlers)]
pivot = df_filtered.pivot_table(values='batsman_runs', index='batsman', columns='bowler', aggfunc='sum')

sns.heatmap(pivot, annot=True, cmap='Reds')
plt.show()
```

```python
# Distribution of runs in different phases of the innings (first 6 overs, middle overs, death overs)
# Define phases
def get_phase(over):
    if over < 6:
        return 'Powerplay'
    elif over < 16:
        return 'Middle'
    else:
        return 'Death'

df['phase'] = df['over'].apply(get_phase)

# Calculate runs per ball by phase
phase_runs = df.groupby('phase')['total_runs'].mean()


phase_runs.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Runs per Ball by Innings Phase')
plt.xlabel('Phase')
plt.ylabel('Runs per Ball')
plt.show()

print(phase_runs)
```

```python
# Strike rate analysis for top batsmen
# Calculate strike rate for top batsmen
for batsman in df.groupby('batsman')['batsman_runs'].sum().nlargest(5).index:
    data = df[df['batsman'] == batsman]
    sr = (data['batsman_runs'].sum() / len(data)) * 100

    print(f"{batsman}: {sr:.1f}")
```

```python
# Simple match runs summary
match_summary = df.groupby(['match_id', 'inning'])['total_runs'].sum().unstack()
# Select only the first two innings (columns 1 and 2) and rename them
match_summary = match_summary[[1, 2]]
match_summary.columns = ['Innings_1', 'Innings_2']

print("\nMatch Runs Summary:")
print(match_summary.head(10))
```
